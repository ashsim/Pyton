```python
import json
import urllib2
from lxml import html
url = "https://www.youtube.com/feed/trending"
response = urllib2.urlopen(url)
page = response.read()
tree = html.document_fromstring(page)
title = tree.cssselect('[class="yt-uix-tile-link yt-ui-ellipsis yt-ui-ellipsis-2 yt-uix-sessionlink      spf-link "] ')
duration=tree.cssselect('[class="video-time"]')
name=tree.cssselect('[class="g-hovercard yt-uix-sessionlink       spf-link "]')
view = tree.cssselect('[class="yt-lockup-meta-info"] ' )
x=[]
for i in view:
    x.append(i.text_content())
xx=str(x)
import re
views = re.findall('\S([0-9]+\S*)\s+views',xx)
data=[]
for i in range(len(title)):
    data.append({'Title' :title[i].get("title"),"URL" : "https://www.youtube.com"+title[i].get('href'),'Duration':duration[i].text_content(),"Username":name[i].text_content(),"Views": views[i]})
from pprint import pprint
pprint(data)
with open('youtube3.json', 'w') as t:
    json.dump(data, t, sort_keys = True, indent = 4)


import unicodecsv as csv
with open('youtube4.csv', 'w') as f:

    writer = csv.writer(f)
    writer.writerow(["Title","URL","Duration","Username","Views"])
    for a in data:
        writer.writerow([a['Title'],a['URL'],a['Duration'],a['Username'],a['Views']])
```
